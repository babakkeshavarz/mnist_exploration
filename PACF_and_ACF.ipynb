```python
from statsmodels.graphics.tsaplots import plot_pacf, plot_acf
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x0 = np.random.randn(1000)
plt.plot(x0)
```

```python
fig, ax = plt.subplots(figsize = (15,5))
plot_pacf(x0 , ax)
```

```python
x1 = [0]
for i in range (1000):
  x = 0.5 * x1[-1]  + 0.1 * np.random.randn()
  x1.append(x)
x1 = np.array(x1)
plt.plot(x1)
```

```python
fig, ax = plt.subplots(figsize = (15,5))
plot_pacf(x1 , ax)
```

```python
x5 = [0 , 0 , 0, 0 , 0]
for i in range (1000):
  x = 0.5 * x5[-1] - 0.1 * x5[-3] + 0.1 * x5[-5] + 0.1 * np.random.randn()
  x5.append(x)
x5 = np.array(x5)
plt.plot(x5)
```

```python
fig, ax = plt.subplots(figsize = (15,5))
plot_pacf(x5 , ax)
```
